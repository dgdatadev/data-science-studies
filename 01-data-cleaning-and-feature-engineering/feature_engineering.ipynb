```python
#Importando o pandas
import pandas as pd
```

```python
#Importando a base de dados tradada
base = pd.read_excel('ChavesClientesTratada.xlsx')
base.head()
```

```python
# Removendo as colunas descenessárias
base = base.drop(['ID','ChaveSituacao','ClassRisco','CatCliente'],axis=1)
base.head()
```

```python
# Importando o OneHotEnconder e transformando as colunas Genero e EstadoCivil em números
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
ohe_transform = ohe.fit_transform(base[['Genero', 'EstadoCivil']])
```

```python
ohe.get_feature_names_out()
```

```python
ohe_transform.toarray()
```

```python
# Criando um dataframe com os dados
df_ohe = pd.DataFrame(ohe_transform.toarray(), columns=ohe.get_feature_names_out())
df_ohe.head()
```

```python
# Incluindo o dataframe da base tradada
base = pd.concat([base, df_ohe], axis=1)
base.head()
```

```python
base.Categoria.values.reshape(-1,1)
```

```python
# Importando o OrdinalEnconder e transformando a coluna Categoria em números
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
oe_transform = oe.fit_transform(base.Categoria.values.reshape(-1,1))
```

```python
oe_transform
```

```python
# Incluindo os dados na base
base['NrCategoria'] = oe_transform
base.head()
```

```python
# Transformando a coluna Risco
oe = OrdinalEncoder(categories=[['C-','C','C+','B-','B','B+','A-','A','A+',]])
oe_transform_risco = oe.fit_transform(base.Risco.values.reshape(-1,1))
oe_transform_risco
```

```python
# Incluindo a coluna transformada na base
base['NrRisco'] = oe_transform_risco
base.head()
```

```python
# Definindo uma função com dados da coluna CatVIP
def define_VIP(valor):
    if valor == 'Alpha' or valor == 'Beta':
        return 1
    else:
        return 0
```

```python
# Incluindo a coluna na base
base['NrVIP'] = base.CatVIP.apply(define_VIP)
base.head()
```

```python
# Removendo as colunas descenessárias
base = base.drop(['Genero','EstadoCivil','Categoria','Risco','CatVIP'],axis=1)
base.head()
```

```python
X = base.drop('Pagamento',axis=1)
y = base.Pagamento

from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X,y)

reg.score(X,y)
```
