```python
from sklearn.model_selection import train_test_split
```

```python
X = base.drop('aluguel',axis=1)
y = base.aluguel
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.head(3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(X_train, y_train)
```

```python
reg.score(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
y_pred = reg.predict(X_test)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
fig, ax= plt.subplots()
ax.scatter(y_test, y_pred)
ax.plot([0,10000],[0,10000])
plt.show()
```

```python
from sklearn import tree
```

```python
regArvore = tree.DecisionTreeRegressor(random_state=42)
```

```python
regArvore = regArvore.fit(X_train, y_train)
```

```python
regArvore.score(X_train, y_train)
```

```python
y_predArvore = regArvore.predict(X_test)
```

```python
mean_absolute_error(y_test, y_predArvore)
```

```python
fig, ax= plt.subplots()
ax.scatter(y_test, y_pred)
ax.scatter(y_test, y_predArvore)
ax.plot([0,10000],[0,10000])
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_predArvore)
```

```python
from sklearn.svm import SVR
```

```python
reg.SVR = SVR()
reg.SVR = reg.SVR.fit(X_train, y_train)
y_predSVR = reg.SVR.predict(X_test)
mean_absolute_error(y_test, y_predSVR)
```

```python
fig, ax= plt.subplots()
ax.scatter(y_test, y_pred)
ax.scatter(y_test, y_predArvore)
ax.scatter(y_test, y_predSVR)
ax.plot([0,10000],[0,10000])
plt.show()
```
