```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,]
y = [2,3,4,3]

plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,]
y = [2,3,4,3]
plt.plot( x, y, label='dados')
plt.ylabel("Eixo y")
plt.xlabel("Eixo x")
plt.title("Titulo do Gráfico")
plt.xticks([0,2,4,6,8,9])
plt.yticks([-1,3,5,9,11])
plt.legend()
plt.show()
```

```python
plt.plot(x,y,linestyle='--',color='y',lw=5.0)
plt.show()
```

```python
plt.scatter(x,y,color='r')
plt.show()
```

```python
plt.bar(x,y,color='g')
plt.show()
```

```python
plt.scatter(x,y,color='r',marker='s')
plt.show()
```

```python
plt.plot(x,y)
plt.axis('square')
plt.show()
```

```python
valoresx = [1, 2, 3, 4]
valoresy = [1, 4, 2, 3]

figura = plt.figure(figsize=(20,3))
figura.suptitle('Titulo Geral')

figura.add_subplot(131)
plt.plot(valoresx,valoresy, label='Um dado qualquer')
plt.legend()
plt.title('Gráfico 1')

figura.add_subplot(132)
plt.scatter(valoresx,valoresy)
plt.title('Gráfico 2')

figura.add_subplot(133)
plt.bar(valoresx,valoresy)
plt.title('Gráfico 3')

plt.savefig('graficos.png')

plt.show(figura)
```
