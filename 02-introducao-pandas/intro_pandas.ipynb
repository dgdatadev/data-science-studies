```python
vendas_df.loc[:, 'Imposto'] = 0
display(vendas_df)
```

```python
vendas_dez_df = pd.read_excel("/content/drive/MyDrive/Google Colab Notebooks/Introdução ao Pandas/Vendas - Dez.xlsx")
display(vendas_dez_df)
```

```python
# Mesclando duas tabelas
vendas_df = pd.concat([vendas_df, vendas_dez_df])
display(vendas_df)
```

```python
# Removendo uma coluna
vendas_df = vendas_df.drop("Imposto", axis=1)
display(vendas_df)
```

```python
vendas_df = vendas_df.drop(0, axis=0)
display(vendas_df)
```

```python
# Maneiras de remover celulas vazias (
#   vendas_df = vendas_df.dropna(how='all', axis=1)
#   vendas_df = vendas_df.dropna()
# )

#Ou preencher com a média dos valores na mesma coluan
vendas_df['Comissão'] = vendas_df['Comissão'].fillna(vendas_df['Comissão'].mean())
display(vendas_df)
```

```python
vendas_df = vendas_df.ffill()
display(vendas_df)
```

```python
transacoes_lojas = vendas_df['ID Loja'].value_counts()
display(transacoes_lojas)
```

```python
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturamento_produto)
```
