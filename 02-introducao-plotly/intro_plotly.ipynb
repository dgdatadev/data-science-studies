```python
import plotly.express as px
```

```python
dados_x = ['2018', '2019', '2020', '2021']
dados_y = [10, 20, 4, 35]

fig = px.line(x=dados_x, y=dados_y, title= 'Vendas X Ano', width=600, height=300, line_shape= 'spline')
fig.update_yaxes(title='Vendas', title_font_color='red')
fig.update_xaxes(title='Ano', title_font_color='blue')
fig.show()
```

```python
dados_x = ['2018', '2019', '2020', '2021']
dados_y = [10, 20, 4, 35]

fig = px.pie(names=dados_x, values=dados_y, title= 'Vendas X Ano', width=400, height=400)
fig.show()
```

```python
dados_x = ['2018', '2019', '2020', '2021']
dados_y = [10, 20, 4, 35]

fig = px.bar(x=dados_x, y=dados_y, title= 'Vendas X Ano', width=500 , height=400)
fig.show()
```

```python
dados_x = [1, 4, 6, 7, 8, 4, 3, 2, 1, 5]
dados_y = [10, 20, 5, 35, 2, 3, 40, 25, 16, 27]

fig = px.scatter(x=dados_x, y=dados_y, title= 'Vendas X Ano', width=500 , height=400)
fig.show()
```
