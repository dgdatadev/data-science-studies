```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_videos = pd.read_excel('/content/drive/MyDrive/Google Colab Notebooks/Introdução ao Seaborn/videosYT.xlsx','videos')
display(df_videos)
```

```python
sns.set_theme(style='darkgrid')
sns.scatterplot(data=df_videos, x='Nº de Views', y='Nº de Likes', hue='Responsável', palette = ['red', 'black']);
```

```python
sns.relplot(data=df_videos, x='Nº de Views', y='Nº de Likes', hue='Categoria', col='Responsável');
```

```python
df_inscritos = pd.read_excel('/content/drive/MyDrive/Google Colab Notebooks/Introdução ao Seaborn/videosYT.xlsx','Inscritos')
display(df_inscritos)
```

```python
sns.lineplot(data=df_inscritos, x='Mês/Ano', y='Inscritos', color = 'red');
```

```python
sns.displot(data=df_videos, x='Nº de Views', hue='Responsável', col='Responsável', rug=True);
```

```python
sns.regplot(data=df_videos, x='Nº de Views', y='Nº de Likes');
```

```python
sns.lmplot(data=df_videos, x='Nº de Views', y='Nº de Likes', hue='Responsável', markers=['o','x']);
```
